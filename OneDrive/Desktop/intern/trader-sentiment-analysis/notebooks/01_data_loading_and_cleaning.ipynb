# 01 - Data Loading and Cleaning
Load raw datasets, clean fear-greed and trade records, and save merged data.

```python
from pathlib import Path
import pandas as pd

from src.data_loader import load_raw_datasets
from src.preprocessing import preprocess_fear_greed, preprocess_trades, merge_datasets, save_processed_dataset
```

```python
project_root = Path('..').resolve()
raw_dir = project_root / 'data' / 'raw'
processed_dir = project_root / 'data' / 'processed'

fg_raw, trades_raw = load_raw_datasets(
    fear_greed_path=raw_dir / 'fear_greed_index.csv',
    trades_path=raw_dir / 'hyperliquid_trades.csv'
)

fg_clean, fg_diag = preprocess_fear_greed(fg_raw)
trades_clean, quarantine = preprocess_trades(trades_raw)
merged = merge_datasets(trades_clean, fg_clean)

save_processed_dataset(merged, processed_dir / 'merged_dataset.csv')
fg_diag.to_csv(processed_dir / 'fear_greed_diagnostics.csv', index=False)
quarantine.to_csv(processed_dir / 'quarantined_records.csv', index=False)

print('Fear/Greed rows:', len(fg_clean))
print('Trades rows cleaned:', len(trades_clean))
print('Quarantined rows:', len(quarantine))
print('Merged rows:', len(merged))
```
